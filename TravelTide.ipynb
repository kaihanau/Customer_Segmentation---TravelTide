```python
query = """
with
  sessions_in_timeframe as (
    select
      s.*,
       (extract(epoch from s.session_end) - extract(epoch from s.session_start)) session_duration,
      u.birthdate,
      u.gender,
      u.married,
      u.has_children,
      u.home_country,
      u.home_city,
      u.home_airport,
      u.home_airport_lat,
      u.home_airport_lon,
      u.sign_up_date,
      f.origin_airport,
      f.destination,
      f.destination_airport,
      f.seats,
      f.return_flight_booked,
      f.departure_time,
      f.return_time,
      f.checked_bags,
      f.trip_airline,
      f.destination_airport_lat,
      f.destination_airport_lon,
      f.base_fare_usd,
      extract(epoch from (f.departure_time - s.session_end)) time_before_travel,
      case when s.flight_discount_amount is not null then (f.base_fare_usd * (1 - s.flight_discount_amount)) else f.base_fare_usd end actual_ticket_price,
      h.hotel_name,
      case when h.nights < 0 then 1 else nights end nights,
      h.rooms,
      h.check_in_time,
      h.check_out_time,
      h.hotel_per_room_usd as hotel_per_room_night_usd,
      extract(epoch from (h.check_in_time - s.session_end)) time_before_hotel_stay,
      case when s.hotel_discount_amount is not null then (h.hotel_per_room_usd * (1 - s.hotel_discount_amount)) else h.hotel_per_room_usd end actual_hotel_per_room_usd
    from
      sessions s
      left join users u on s.user_id = u.user_id
      left join hotels h on h.trip_id = s.trip_id
      left join flights f on s.trip_id = f.trip_id
    where
      s.session_start >= '2023-01-04'
  ),

  filtered_users as (
    select
      count(session_id) num_sessions,
      user_id
    from
      sessions
    where
      session_start > '2023-01-04'
    group by
      user_id
    having
      count(session_id) > 7
  ),

  base_sessions as (
    select
      *,
      case when (flight_discount = true or hotel_discount = true) then 1
      else 0
      end discount_aplied
    from
      sessions_in_timeframe
    where
      user_id in (
        select
          user_id
        from
          filtered_users
      )
  ),

   cancelled_trips as ( select * from base_sessions
  where cancellation = true),

  sessions_details_per_user as
  (
  select
  user_id,
  count(session_id) num_sessions,
  sum(page_clicks) num_clicks,
  avg(page_clicks) average_num_clicks,
  sum(session_duration) total_session_time,
  avg(session_duration) average_duration,
  sum(discount_aplied) num_disc_sessions

from
base_sessions

group by user_id
),

trips_per_user as
(
select
  user_id,
  count(distinct trip_id) num_completed_trips,
  sum(discount_aplied) num_disc_trips,
  sum(discount_aplied)/count(distinct trip_id)::numeric perc_disc_trips
from
base_sessions
where trip_id not in (select trip_id from cancelled_trips)
group by user_id
),

flights_details_per_user as
(
select
  user_id,
  sum(actual_ticket_price * seats) total_flight_price_paid,
  avg(actual_ticket_price) avg_flight_price_paid,
  sum(seats) total_num_seats,
  avg(seats) avg_num_seats,
  avg(checked_bags) avg_num_checked_bags,
  count(actual_ticket_price) total_num_flights,
  count(flight_discount_amount) total_num_disc_flights,
  round(cast(count(flight_discount_amount)/count(actual_ticket_price)::float * 100 as numeric), 2) perc_disc_flights,
  avg(flight_discount_amount) avg_flight_discount,
  avg(time_before_travel)/(3600 * 24) avg_time_before_travel
from
base_sessions
where trip_id not in (select trip_id from cancelled_trips)
and base_fare_usd is not null
group by user_id
),

hotel_bookings_per_user as
(
select
user_id,
  sum(actual_hotel_per_room_usd * nights * rooms) total_hotel_night_paid,
  sum(nights) total_num_nights,
  sum(rooms) total_num_rooms,
  avg(nights) avg_num_nights,
  avg(rooms) avg_num_rooms,
  count(hotel_per_room_night_usd) total_num_hotel_bookings,
  count(hotel_discount_amount) total_num_disc_hotel_bookings,
  round(cast(count(hotel_discount_amount)/count(hotel_per_room_night_usd)::float * 100 as numeric), 2) perc_disc_bookings,
  avg(actual_hotel_per_room_usd) average_room_price,
  coalesce(avg(hotel_discount_amount), 0) avg_hotel_discount,
  avg(time_before_hotel_stay)/(3600 * 24) avg_time_before_hotel_stay
from
base_sessions
where trip_id not in (select trip_id from cancelled_trips)
and hotel_per_room_night_usd is not null
group by user_id
order by count(hotel_discount_amount) desc
),

users_impr as
(
select user_id,	birthdate, round((current_date - birthdate)/365.0) user_age,
 (current_date - sign_up_date) days_with_company,
 gender,
 married,
 has_children,
 home_country,
 home_city,
 home_airport,
 home_airport_lat,
 home_airport_lon,
 sign_up_date
from users
),

users_cte as
(
  select user_id,	birthdate, user_age,
case
when user_age between 16 and 24 then '16 - 24'
when user_age between 25 and 34 then '25 - 34'
when user_age between 35 and 44 then '35 - 44'
when user_age between 45 and 54 then '45 - 54'
when user_age between 55 and 64 then '55 - 64'
else 'over 64'
end age_group,
 (current_date - sign_up_date) days_with_company,
 gender,
 married,
 has_children,
 home_country,
 home_city,
 home_airport,
 home_airport_lat,
 home_airport_lon,
 sign_up_date
from users_impr
  ),

recency_table as
(
  select user_id,
  date_part('day', '2023-08-28' - max(session_end)) recency
  from base_sessions
  group by user_id
),

users_base as
(
select
users_cte.*,
num_sessions,
num_clicks,
average_num_clicks,
total_session_time,
average_duration,
num_disc_sessions,
coalesce(total_flight_price_paid, 0) total_flight_price_paid,
coalesce(avg_flight_price_paid, 0) avg_flight_price_paid,
coalesce(total_num_seats, 0) total_num_seats,
coalesce(avg_num_checked_bags, 0) avg_num_checked_bags,
coalesce(avg_num_seats, 0) avg_num_seats,
coalesce(total_num_flights, 0) total_num_flights,
coalesce(total_num_disc_flights, 0) total_num_disc_flights,
coalesce(perc_disc_flights, 0) perc_disc_flights,
coalesce(avg_flight_discount, 0) avg_flight_discount,
coalesce(avg_time_before_travel, 0) avg_time_before_travel,
coalesce(total_hotel_night_paid, 0)total_hotel_night_paid,
coalesce(total_num_nights, 0) total_num_nights,
coalesce(avg_num_nights, 0) avg_num_nights,
coalesce(avg_num_rooms, 0) avg_num_rooms,
coalesce(total_num_rooms, 0) total_num_rooms,
coalesce(total_num_hotel_bookings, 0) total_num_hotel_bookings,
coalesce(total_num_disc_hotel_bookings, 0) total_num_disc_hotel_bookings,
coalesce(perc_disc_bookings, 0) perc_disc_bookings,
coalesce(average_room_price, 0) average_room_price,
coalesce(avg_time_before_hotel_stay, 0) avg_time_before_hotel_stay,
coalesce(recency, 0) days_from_last_session,
(coalesce(total_flight_price_paid, 0) + coalesce(total_hotel_night_paid, 0))total_spend,
coalesce(num_completed_trips, 0) total_num_trips,
coalesce(num_disc_trips, 0) num_disc_trips,
coalesce(perc_disc_trips, 0) perc_disc_trips
from users_cte
inner join
sessions_details_per_user
on users_cte.user_id = sessions_details_per_user.user_id
left join trips_per_user
on trips_per_user.user_id = users_cte.user_id
left join flights_details_per_user
on flights_details_per_user.user_id = users_cte.user_id
left join hotel_bookings_per_user
on users_cte.user_id = hotel_bookings_per_user.user_id
left join recency_table
on sessions_details_per_user.user_id = recency_table.user_id
),

users_base_rfm_score as
(
select *,
  ntile(3) over (order by days_from_last_session desc) recency,
  ntile(3) over (order by total_num_trips) frequency,
  ntile(3) over (order by total_spend) monetary
  from users_base
),

users_base_overall_rfm_score as
(
select *,
concat(recency, frequency, monetary)::integer rfm
from users_base_rfm_score
),

segmentation as
(
select *,
case
 when (rfm in (311, 312, 311) and total_num_trips > 0)
 then 'new travelers'


 when (rfm in (111, 121, 131, 122, 133, 113, 112, 132) and total_num_trips > 0)
 then 'lost travelers'


 when (total_num_trips > 0 and rfm in (212, 123, 221, 211)) then 'regular travelers'

 when (total_num_trips > 0 and rfm in (223, 222, 213, 322, 231, 232, 321, 331)) then 'loyal travelers'

 when (total_num_trips > 0 and rfm in (333, 332, 313, 323, 233)) then 'top  travelers'

 when total_num_trips = 0 then 'potential travelers'
 end cust_category
 from users_base_overall_rfm_score
)

select *,
coalesce
(
case
when cust_category = 'top  travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights > 0 then 'top business traveller - seat upgrade'
when cust_category = 'top  travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights = 0 then 'top business traveller - room upgrade'
when cust_category = 'top  travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights > 0 then 'top family traveler - free plane ticket cancellation'
when cust_category = 'top  travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights = 0 then 'top family traveler - free hotel cancellation'
when cust_category = 'top  travelers' and user_age > 64 then 'free theater tickets'
when cust_category = 'top  travelers' then 'top traveler - free 1 hotel night'
else null
end,
case
when cust_category = 'loyal travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights > 0 then 'loyal business traveller - airport lounge access'
when cust_category = 'loyal travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights = 0 then 'loyal business traveller - free taxi service'
when cust_category = 'loyal travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights > 0 then 'loyal family traveler - up to 2 free bags'
when cust_category = 'loyal travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights = 0 then 'loyal family traveler - free room service'
when cust_category = 'loyal travelers' and user_age > 64 then 'loyal senior traveler - free gallery/museum tickets'
when cust_category = 'loyal travelers' then 'loyal traveler - free spa visit'
else null
end,
case
when cust_category = 'regular travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights > 0 then 'regular business traveller - free airport pickup'
when cust_category = 'regular travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights = 0 then 'regular business traveller - free dinner'
when cust_category = 'regular travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights > 0 then 'regular family traveler - 1 free bags'
when cust_category = 'regular travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights = 0 then 'loyal family traveler - free meal for family'
when cust_category = 'regular travelers' and user_age > 64 then 'loyal senior traveler - airport/hotel assistance'
when cust_category = 'regular travelers' then 'regular traveler - complimentary meal'
else null
end,
case
when cust_category = 'new travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights > 0 then 'new business traveller - priority check-in'
when cust_category = 'new travelers' and avg_num_checked_bags <= 1 and avg_num_nights <= 3
and  total_num_flights = 0 then 'new business traveller - free bottle of wine'
when cust_category = 'new travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights > 0 then 'new family traveler - free tickets to a family show'
when cust_category = 'new travelers' and user_age < 45 and married = true and has_children = true
and total_num_flights = 0 then 'new family traveler - a gift bag'
when cust_category = 'new travelers' and user_age > 64 then 'new senior traveler - complimentary meal'
when cust_category = 'new travelers' then 'new traveler - 10% discount'
else null
end,
case
when cust_category = 'lost travelers' then '20% discount' else null end,
case when cust_category = 'potential travelers' then '25% discount'else null end
) perks
from segmentation
"""







```
